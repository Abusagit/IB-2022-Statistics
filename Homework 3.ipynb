# Задание 2

Функция правдоподобия будет иметь вид:

$$L(\mu) = \prod_{i=1}^n\frac{1}{2}\exp\{-|X_i - \mu|\} = 2^{-n} \exp\{-\sum_{i=1}^n|X_i - \mu|\}$$

Логарифм функции правдоподобия:

$$log L(\mu) = -n\ln(2) - \sum_{i=1}^n|X_i - \mu|$$

Значение зависит от второго слагаемого, его и необходимо минимизировать, чтобы максимизировать функцию.

Можно сказать, что функция от двух параметров $f(X_i, \mu) = |X_i - \mu|$ будет иметь различные производные в зависимости от значения $X_i$, а именно:

\begin{equation}
    \frac{\partial f}{\partial \mu} = 
    \begin{cases}
    -1, & X_i \geq \mu, \\
    1, & X_i < \mu, \\
    undefined, & X_i = \mu
    \end{cases}
\end{equation}

Можем поспользоваться данным знанием о производной и написать уравнение правдоподобия:

\begin{align}
    \frac{\partial log L}{\partial \mu} &= \frac{\partial \{ - \sum_{i=1}^n|X_i - \mu| \}}{\partial \mu} = \\
    &= - \sum_{i=1}^n \frac{\partial f}{\partial \mu} = \\
    &= \sum_{i=1}^n sign(X_i - \mu)
\end{align}
Здесь уже нивелируется неопределенность в точке 0, так как $sign(0) = 0$, а в других случаях - все тем более сходится.

Таким образом, если $n$ - нечетное, то:

$$\frac{\partial log L}{\partial \mu} = 0 \iff \mu = \overline{z_{\frac{1}{2}}},$$
То есть, когда $\mu$ - медиана, $\frac{n-1}{2}$ значений будут больше медианы, и $\frac{n - 1}{2}$ - меньше, и сама медиана будет представлена в выборке, что занулит слагаемое суммы.


Если $n$ - четное, то мы все еще можем минимизировать функцию правдоподобия, подобрав в качестве значения $\mu$ необходимый член вариационного ряда - $X_{[\frac{n}{2}]}$ или $X_{[\frac{n+1}{2}]}$ - а так как для выборки с четным количеством слагаемых медиану можно взять как $X_{[\frac{n+1}{2}]}$ (пройдено на лекции), то выходит, что в обоих случаях ОМП для параметра $\mu$ есть медиана, она же квантиль уровня 1/2:

$$\hat{\mu} = \overline{z_{\frac{1}{2}}}$$

# Задание 4


Для начала __покажем асимптотическую несмещенность__ при помощи среднеквадратического отклонения.

Напомним формулу среднеквадратического отклонения:
$$\E[(\hat{\theta_n} - \theta_0)^2] = \D[\hat{\theta_n}] + \mathbf{b}^2(\hat{\theta_n})$$

Оценка асимптотически несмещена, когда слагаемое $\mathbf{b}^2(\hat{\theta_n}) \xrightarrow[n\to\infty]{}0$. Покажем это, а затем смоделируем.

С практики мы знаем, что $p_{X_{[n]}}(t) = \frac{nx^{n-1}}{\theta_0^n}$. Запишем необходимые значения величин.

$$\E[\hat{\theta_n}] = \frac{n}{n+1}\theta_0$$
$$\mathbf{b}^2(\hat{\theta_n}) = (\E[\hat{\theta_n}] - \theta_0)^2 = \l(\frac{n}{n+1}\theta_0 - \theta_0\r)^2 = \frac{1}{(n+1)^2}\theta_0^2\xrightarrow[n\to+\infty]{}0\implies$$

$\implies$ оценка асимптотически несмещена, но мы продемонстрируем это прямым моделированием. Как видно, для данного акта не нужно моделировать распределения - нам необходимы лишь набор значений $theta_0$ и набор из значений $n$ для подсчета квадрата смещения для данного $\theta_0$ и $n$.

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm_notebook as tqdm

plt.style.use('seaborn')
SIZE = 22
# Выставим дефолтные параметры плоттинга
plt.rc('font', size=SIZE)
plt.rc('axes', titlesize=SIZE)
plt.rc('axes', labelsize=SIZE)
plt.rc('xtick', labelsize=SIZE)
plt.rc('ytick', labelsize=SIZE)
plt.rc('legend', fontsize=SIZE)
plt.rc('figure', titlesize=SIZE)
```

```python
def bias_convergence(theta_0, sample_sizes):
    # функция подсчета квадрата отклонения для данного theta и n
    bias = np.vectorize(lambda n: theta_0 ** 2 / (n + 1) ** 2) 
    
    # массив, содержащий значение квадрата отклонения для каждого n и для данного theta
    return bias(sample_sizes)

def convergence_of_function(func, theta_array):
    # двумерный массив, в котором строка - значения квадрата отклонения для значения theta_0
    convergence_array = []
    
    for theta in theta_array:
        convergence_array.append(func(theta, sample_sizes))
        
    
    return np.array(convergence_array)


def plot_convergence(data, theta_array, characteristic):
    """
    Функция для визуализации результатов
    
    """
    
    plt.figure(figsize=(20, 12))
    
    for i, theta in zip(range(data.shape[0]), theta_array):
        sns.lineplot(y=data[i], x=sample_sizes, label=f"Theta_0 = {theta}")
    
    plt.ylabel(f'Sample {characteristic}') 
    plt.xlabel('Sample size') 
    plt.xticks(ticks=sample_sizes)
    plt.title(f"Convergence of {characteristic}")
    plt.legend(title="")
    plt.ylim([0, np.max(data)])
    plt.xlim([sample_sizes[0], sample_sizes[-1]])

    # шкала в логарифмической системе лучше отразит поведение на малых значениях выборки
    plt.xscale("log")
    plt.show()
```

```python
# Массив с различными значениями n
sample_sizes = np.concatenate([np.arange(5, 105, 5), 10 * np.arange(10, 110, 10), 100 * np.arange(10, 210, 10)])
thetas_array = [1, 5, 10, 15, 20]
```

```python
bias_data = convergence_of_function(bias_convergence, thetas_array)
plot_convergence(bias_data, thetas_array, "bias^2")
```

Таким образом, показав, что bias^2 стремится к 0, мы продемонстрировали асимптотическую несмещенность оценки

Теперь __исследуем скорость сходимости оценки $\hat{\theta}_n$ к истинному значению параметра $\theta$__. Для этого воспользуемся среднеквадратическим отклонением. Мы нашли слагаемое квадрата отклонения, теперь для начала найдем дисперсию оценки:

\begin{align}
    \D[\hat{\theta}_n] &= \E[(\hat{\theta}_n)^2] - (\E[\hat{\theta}_n])^2 = \\
    &= \frac{n}{\theta_0^n}\int_0^{\theta_0}t^{n+1}dt - \frac{n^2}{(n+1)^2}\theta_0^2 = \\
    &= \frac{n}{n+2}\theta_0^2 - \frac{n^2}{(n+1)^2}\theta_0^2 \implies
\end{align}

$\implies$
\begin{align}
    \E[(\hat{\theta_n} - \theta_0)^2] &= \D[\hat{\theta_n}] + \mathbf{b}^2(\hat{\theta_n}) = \\
    &= \frac{n}{n+2}\theta_0^2 - \frac{n^2}{(n+1)^2}\theta_0^2 + \frac{1}{(n+1)^2}\theta_0^2 = \\
    &= \frac{n}{n+2}\theta_0^2 + \frac{1 - n^2}{(n+1)^2}\theta_0^2 = \\
    &= \l[\frac{n}{n+2} + \frac{1-n}{n+1}\r]\theta_0^2 = \\
    &= \frac{2}{n^2 + 3n +2}\theta_0^2 = \mathcal{O}\l(\frac{1}{n^2}\r)
\end{align}


Таким образом, оценка имеет скорость сходимости порядка $\frac{1}{n^2}$ - что является сверхэффективной оценкой. Теперь покажем, как быстро убывает среднеквадратическое отклонение:

```python
# Функция идентична bias_convergence, изменена лишь формула - здесь - 
# - среднеквадратическое отклонение
def mse_convergence(theta_0, sample_sizes):
    mse_vectorized = np.vectorize(lambda n: (2 * theta_0 ** 2) / (n ** 2 + 3 * n + 2))
    
    return mse_vectorized(sample_sizes)
```

Так как оценка сверхэффективная, можем сократить массив с размерами выборки - уменьшим его до максимального значения 1000:

```python
sample_sizes = np.concatenate([np.arange(5, 105, 5), 10 * np.arange(10, 110, 10)])
```

```python
mse_data = convergence_of_function(mse_convergence, thetas_array)
```

```python
plot_convergence(mse_data, thetas_array, "MSE")
```

Таким образом, мы показали, что скорость сходимости оценки к истинному значению имеет порядок 1/n^2 (графики действительно напоминают гиперболическую кривую)
